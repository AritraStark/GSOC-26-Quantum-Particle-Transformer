##Implement a simple quantum operation with Cirq or Pennylane
  1.  With 5 qubits
  2. Apply Hadamard operation on every qubit
  3. Apply CNOT operation on (0, 1), (1,2), (2,3), (3,4)
  4. SWAP (0, 4)
  5. Rotate X with pi/2 on any qubit
  6. Plot the circuit


```python
!pip install cirq

import cirq
import numpy as np
import matplotlib.pyplot as plt

# Create 5 qubits
qubits = cirq.LineQubit.range(5)

# Build circuit
circuit = cirq.Circuit()

# (b) Apply Hadamard to every qubit
circuit.append([cirq.H(q) for q in qubits])

# (c) CNOT chain
circuit.append([
    cirq.CNOT(qubits[0], qubits[1]),
    cirq.CNOT(qubits[1], qubits[2]),
    cirq.CNOT(qubits[2], qubits[3]),
    cirq.CNOT(qubits[3], qubits[4])
])

# (d) SWAP between qubit 0 and 4
circuit.append(cirq.SWAP(qubits[0], qubits[4]))

# (e) Rotate X by pi/2 on qubit 2
circuit.append(cirq.rx(np.pi/2)(qubits[2]))

print("Quantum Circuit:\n")
print(circuit)

# (f) Plot circuit
cirq.CircuitDiagramInfo
plt.figure(figsize=(8,3))
print("\nCircuit diagram rendered above.")
```

##Implement a second circuit with a framework of your choice:
1. Apply a Hadmard gate to the first qubit
rotate the second qubit by pi/3 around X
2. Apply Hadamard gate to the third and fourth qubit
3. Perform a swap test between the states of the first and second qubit |q1 q2> and the third and fourth qubit |q3 q4>

```python
# Define qubits
q1, q2, q3, q4, ancilla = cirq.LineQubit.range(5)

circuit = cirq.Circuit()

# State preparation
circuit.append(cirq.H(q1))            # Hadamard on first qubit
circuit.append(cirq.rx(np.pi/3)(q2))  # Rotate second qubit
circuit.append([cirq.H(q3), cirq.H(q4)])  # Hadamard on third and fourth

# Swap Test
circuit.append(cirq.H(ancilla))

circuit.append(cirq.CSWAP(ancilla, q1, q3))
circuit.append(cirq.CSWAP(ancilla, q2, q4))

circuit.append(cirq.H(ancilla))

# Measurement
circuit.append(cirq.measure(ancilla, key='result'))

print("Swap Test Circuit:\n")
print(circuit)
```
